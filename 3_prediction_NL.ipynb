```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_NL.csv")
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df["x"] , df["y"] , train_size=0.2 ,random_state=43)
```

```python
from Linear_regression_code import linear_regression
model = linear_regression(number_of_interation = 5000 , alpha=0.01)
```

```python
model.fit(x_train , y_train)
```

```python
prediction = model.predict(x_test)
```

```python

def plot_actual_vs_predicted(y_true, y_pred):
    """Plots actual vs. predicted values."""
    plt.figure()
    sns.scatterplot(x=y_true, y=y_pred)
    plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], '--', color='red') #diagonal line
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.title("Actual vs. Predicted Values")
    plt.show()
```

```python
m = model.m
c = model.c

y_line = m*df["x"] + c
```

```python
fig ,ax = plt.subplots(nrows=1 , ncols=2, figsize= (14 , 6))
sns.scatterplot(x = df["x"], y = df["y"] , ax=ax[0])
sns.scatterplot(x = df["x"], y = df["y"] , ax=ax[1])
sns.lineplot(x = df["x"], y = y_line , ax=ax[1] , color = "red")
ax[0].set_title("Data relation")
ax[1].set_title("Line fitting after learning");
```

```python
plot_actual_vs_predicted(y_pred=prediction, y_true=y_test)
```
